```python
import scipy.stats as ss
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
df = pd.read_csv("/data/insurance.csv")
```

```python

df.charges.plot.hist(bins = 30)
```

```python
a = ss.norm.rvs(df.charges.mean(), df.charges.std(), 10000)
plt.hist(a, alpha = 0.5, density=True, bins = 30)
df.charges.plot.hist(bins = 30, density = True, alpha = 0.8)

```

```python
ss.normaltest(df.charges)
```

```python
ss.normaltest(a)
```

```python
p = np.linspace(0.01, 0.99, 100)

plt.figure(figsize=(10,5))
plt.subplot(121)

plt.plot(p, p/(1-p))
plt.xlabel("$p$", fontsize = 20)
plt.ylabel(r"$\frac{p}{1-p}$", fontsize = 20)

plt.subplot(122)
plt.plot(p, np.log(p/(1-p)))
plt.xlabel("$p$", fontsize = 20)
plt.ylabel(r"$log\frac{p}{1-p}$", fontsize = 20)
plt.tight_layout()
```
