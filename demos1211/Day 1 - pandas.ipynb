Find top 10 movies based on highest average rating. Consider only those movies that have got 100 or more ratings. Show movieId, title, and avgRating.  

```python
pd.merge?
```

```python
movies = pd.read_csv("/data/ml-latest-small/movies.csv")
movies.info()
```

```python
movies.head()
```

```python
ratings = pd.read_csv("/data/ml-latest-small/ratings.csv")
ratings.info()
```

```python
ratings.head()
```

```python
rating_agg = ratings.groupby("movieId").rating.agg([np.mean, len]).reset_index().query("len>=100")
rating_agg.head()
```

```python
pd.merge(movies, rating_agg, on = "movieId")[["movieId", "title", "mean"]]\
.sort_values("mean", ascending = False).iloc[:10, :]
```
