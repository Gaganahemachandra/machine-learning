```python
import pandas as pd
from sklearn import *
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/abulbasar/data/master/kaggle-houseprice/data_combined_cleaned.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
df = df.drop(columns=["Id"])
```

```python
target = "SalesPrice"
y = np.log(df[target])
X = df.drop(columns=[target])
X_dummy = pd.get_dummies(X, drop_first=True)
X_train, X_test, y_train, y_test = model_selection\
.train_test_split(X_dummy.values, y
                , test_size = 0.3, random_state = 1)

pipe = pipeline.Pipeline([
    ("poly", preprocessing.PolynomialFeatures(degree=1
                                    , include_bias=False)),
    ("scaler", preprocessing.StandardScaler()),
    ("est", linear_model.LinearRegression())
])

pipe.fit(X_train, y_train)

y_train_pred = pipe.predict(X_train)
y_test_pred = pipe.predict(X_test)

rmse_test = metrics.mean_squared_error(y_test, y_test_pred) ** 0.5
rmse_train = metrics.mean_squared_error(y_train, y_train_pred) ** 0.5

r2_test = metrics.r2_score(y_test, y_test_pred)
r2_train = metrics.r2_score(y_train, y_train_pred)

print("rmse of training", rmse_train)
print("rmse of test", rmse_test)
print("r2 of training", r2_train)
print("r2 of test", r2_test)

est = pipe.steps[-1][-1]
```

```python
est.intercept_
```

```python
summary = pd.DataFrame(dict(coeff=est.coef_
                , feature=X_dummy.columns))

summary.head()
```

```python
summary["abs_coeff"] = np.abs(summary.coeff)
summary.head()
```

```python
summary.sort_values("abs_coeff", ascending=False)
```

```python
summary.coeff.plot.bar()
plt.xticks([])
plt.title("Bar plot of the coefficient")
plt.xlabel("Features")
plt.ylabel("Coefficient")
```

```python
result = pd.DataFrame(dict(actual = y_test
                , prediction = y_test_pred))
result["error"] = result.prediction - result.actual
result["error_abs"] = np.abs(result.error)
result.head()
```

```python
result.error.hist(bins = 50, density = True)
plt.title("Histogram of errors")
plt.xlabel("Error")
plt.ylabel("Frequency (normed)")
```

```python
result.sort_values("error_abs", ascending=False)
```

```python
target = "SalesPrice"
y = np.log(df[target])
X = df.drop(columns=[target])
X = pd.get_dummies(X, drop_first=True).values


pipe = pipeline.Pipeline([
    ("poly", preprocessing.PolynomialFeatures(degree=1
                                    , include_bias=False)),
    ("scaler", preprocessing.StandardScaler()),
    ("est", linear_model.ElasticNet(alpha = 0.01, l1_ratio = 0.5
                            , random_state=1))
])

scores = model_selection.cross_val_score(pipe, 
                X, y, cv = 5, scoring = "r2")

scores
```

```python
np.mean(scores)
```

```python
param_grid = {
    "est__alpha": 10 ** np.linspace(-2, 1, 10), 
    "est__l1_ratio": np.linspace(0.0001, 1, 10)
}


gsearch = model_selection.GridSearchCV(pipe, param_grid, verbose=True
        , n_jobs=8, scoring = "r2", cv = 5)

gsearch
```

```python
gsearch.fit(X, y)
```
